```python
import nltk
from nltk import word_tokenize, pos_tag, ne_chunk
from nltk.chunk import tree2conlltags
from nltk.classify import NaiveBayesClassifier
from nltk.metrics import precision
from nltk.probability import FreqDist
from nltk.util import bigrams, trigrams
from nltk.sem import Expression

news = """
Microsoft announced a new AI platform in Chennai.
The company expects the technology to improve business productivity.
"""

tokens = word_tokenize(news)
pos_tags = pos_tag(tokens)

grammar = r"""
  NP: {<NNP>+}
  VP: {<VBD>}
"""
cp = nltk.RegexpParser(grammar)
parse_tree = cp.parse(pos_tags)

subject = "Microsoft"
verb = "announced"
obj = "AI platform"

print("Parsing:")
print("Subject:", subject)
print("Verb:", verb)
print("Object:", obj)

print("\nNamed Entities:")
entities = ne_chunk(pos_tags)

for subtree in entities:
    if hasattr(subtree, "label"):
        entity_name = " ".join([word for word, tag in subtree.leaves()])
        if subtree.label() == "ORGANIZATION":
            print("Organization:", entity_name)
        elif subtree.label() == "PERSON":
            print("Person:", entity_name)
        elif subtree.label() == "GPE":
            print("Location:", entity_name)

train_data = [
    ({"AI": True, "technology": True}, "Technology"),
    ({"software": True, "computer": True}, "Technology"),
    ({"cricket": True, "match": True}, "Sports"),
    ({"election": True, "government": True}, "Politics"),
    ({"stock": True, "market": True}, "Business")
]

classifier = NaiveBayesClassifier.train(train_data)

features = {
    "AI": "AI" in news,
    "technology": "technology" in news
}

category = classifier.classify(features)

print("\nCategory:")
print(category)

test_set = [
    ({"AI": True, "technology": True}, "Technology"),
    ({"cricket": True, "match": True}, "Sports")
]

accuracy = nltk.classify.accuracy(classifier, test_set)

refset = set([0])
testset = set([0])

prec = precision(refset, testset)

print("\nModel Accuracy:")
print(f"{accuracy*100:.0f}%")

print("\nPrecision:")
print(f"{prec:.2f}")

words = [
    w.lower()
    for w in tokens
    if w.isalpha() and w.lower() not in ["a", "the", "to", "in"]
]

fdist = FreqDist(words)

print("\nWord Frequency:")
for word, freq in fdist.items():
    print(word, ":", freq)

total_words = len(words)

print("\nProbability:")
for word in fdist:
    print(f"P({word}) = {fdist[word]}/{total_words} = {fdist[word]/total_words:.2f}")

expr = Expression.fromstring('technology(x)')

print("\nSemantic Representation:")
print(expr)

print("\nBigrams:")
for bg in bigrams(words):
    print(bg)

print("\nTrigrams:")
for tg in trigrams(words):
    print(tg)
```
