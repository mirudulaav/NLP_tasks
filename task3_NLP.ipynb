```python
# Technology News Analysis using NLP

from collections import Counter
import nltk
import string
import matplotlib.pyplot as plt
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

articles = [
    "Artificial intelligence is transforming healthcare and education systems",
    "AI and machine learning are the future of technology and innovation",
    "Healthcare systems are improving with artificial intelligence solutions",
    "Education technology is growing rapidly with AI based tools"
]

# Combine all articles into one text
text = " ".join(articles)

# 1. Convert to lowercase
text = text.lower()

# 2. Remove punctuation
for p in string.punctuation:
    text = text.replace(p, "")

# 3. Tokenization
tokens = word_tokenize(text)

# Remove Stopwords
stop_words = set(stopwords.words('english'))
filtered_words = [word for word in tokens if word not in stop_words]

print("Tokens:")
print(tokens)

print("\nFiltered Words:")
print(filtered_words)

# 4. Word Frequency Distribution
freq = Counter(filtered_words)

print("\nWord Frequency Distribution:")
print(freq)

# 5. Vocabulary (Unique Words)
vocabulary = set(filtered_words)

print("\nVocabulary:")
print(vocabulary)
print("Vocabulary Size:", len(vocabulary))

# 6. Top 5 Most Frequent Words
print("\nTop 5 Frequent Words:")
top5 = freq.most_common(5)

for word, count in top5:
    print(word, ":", count)

# 7. Type Token Ratio (TTR)
total_tokens = len(filtered_words)
ttr = len(vocabulary) / total_tokens

print("\nTotal Tokens:", total_tokens)
print("Type-Token Ratio (TTR):", round(ttr, 2))

# 8. Dominant Topic Analysis
topics = {
    "Artificial Intelligence": ["ai", "artificial", "intelligence", "machine", "learning"],
    "Healthcare": ["healthcare", "systems", "solutions"],
    "Education": ["education"],
    "Technology": ["technology", "innovation", "tools"]
}

topic_scores = {}

for topic, keywords in topics.items():
    score = sum(freq[word] for word in keywords if word in freq)
    topic_scores[topic] = score

print("\nTopic Scores:")
for topic, score in topic_scores.items():
    print(topic, ":", score)

dominant_topic = max(topic_scores, key=topic_scores.get)
print("\nMost Dominant Topic:", dominant_topic)

# 9. Bar Chart Visualization
words = list(freq.keys())
counts = list(freq.values())

plt.figure(figsize=(10, 5))
plt.bar(words, counts)
plt.title("Technology News Word Frequency Analysis")
plt.xlabel("Words")
plt.ylabel("Frequency")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
